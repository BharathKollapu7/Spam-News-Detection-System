# Import Libraries

```python
import pandas as pd
import numpy as np
```

### Import datasets

```python
True_news = pd.read_csv(r'C:\Users\bhara\Desktop\Pythonset\True.csv')
Fake_news = pd.read_csv(r'C:\Users\bhara\Desktop\Pythonset\Fake.csv')\

True_news.head()
Fake_news.head()
```

### Assign labels to recognize as TRUE/FAKE news

```python
True_news['label'] = 0
Fake_news['label'] = 1
```

```python
True_news.head()
Fake_news.head()
```

### Combining the datasets

```python
dataset1 = True_news[['text','label']]
dataset2 = Fake_news[['text','label']]

dataset = pd.concat([dataset1,dataset2])
```

```python
dataset
```

```python
dataset.isnull().sum()
```

### Shuffling Dataset

```python
dataset = dataset.sample(frac=1)
dataset
```

# NLP

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
ps = WordNetLemmatizer()
stopwords = stopwords.words('english')
```

```python
def clean_row(row):
    
    row = row.lower()
    row = re.sub('[^a-zA-Z]',' ',row)
    token = row.split()

    news = [ps.lemmatize(word) for word in token if not word in stopwords]

    cleaned_news = ' '.join(news)
    return cleaned_news
```

```python
dataset['text'] = dataset['text'].apply(lambda x : clean_row(x))
dataset['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features = 50000, lowercase = False, ngram_range=(1,2))
```

```python
x = dataset.iloc[:35000,0]
y = dataset.iloc[:35000, 1]
```

# Train-Test split

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_label, test_label = train_test_split(x,y, test_size = 0.2, random_state = 0)
train_data
```

```python
vec_train_data = vectorizer.fit_transform(train_data)
vec_train_data=vec_train_data.toarray()

vec_test_data = vectorizer.fit_transform(test_data)
vec_test_data = vec_test_data.toarray()
```

```python
training_data = pd.DataFrame(vec_train_data, columns=vectorizer.get_feature_names_out())
testing_data = pd.DataFrame(vec_test_data, columns=vectorizer.get_feature_names_out())
```

```python
training_data
```

# Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
clf.fit(training_data,train_label)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(testing_data)

# Calculate accuracy for test data
test_accuracy = accuracy_score(test_label, y_pred)
print(f"Testing accuracy: {test_accuracy}")

y_pred_train = clf.predict(training_data)

# Calculate accuracy for training data
train_accuracy = accuracy_score(train_label, y_pred_train)
print(f"Training accuracy: {train_accuracy}")
```
